```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of time allocation
plt.figure(figsize=(8, 4))
sns.histplot(df['time'], bins=40, kde=True)
plt.title('Distribution of time allocated')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Extract the hour from the datetime column
df['hour'] = df['time'].dt.hour

# Count the frequency of recordings for each hour and platform
hour_platform_counts = df.groupby(['hour', 'platform']).size().unstack()

# Create a stacked bar plot with colored labels
plt.figure(figsize=(8, 4))
hour_platform_counts.plot(kind='bar', stacked=True, colormap='Set2')
plt.title('Frequency of Recordings Hour Wise')
plt.xlabel('Hours')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.legend(title='Platform')
plt.show()
```

```python
# Extract the year from the datetime column
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month

# Count the frequency of recordings for each year, month, and platform
year_month_platform_counts = df.groupby(['year', 'month', 'platform']).size().unstack()

# Create a stacked bar plot with colored labels
plt.figure(figsize=(8, 4))
year_month_platform_counts.plot(kind='bar', stacked=True, colormap='Set2')
plt.title('Frequency of Recordings by month and year')
plt.xlabel('Year, Month')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.legend(title='Platform')
plt.show()
```

```python
# Group the dataframe by 'type' and count the occurrences
task_frequency = df['type'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 4))
task_frequency.plot(kind='bar')
plt.title('Tasks Frequency')
plt.xlabel('Tasks')
plt.ylabel('Frequency')
plt.xticks(rotation=20)
plt.show()
```

```python
cross_tab = pd.crosstab(df['browser_event'], df['type'])

# Plot the cross-tabulation as a stacked bar chart
cross_tab.plot(kind='bar', stacked=True, figsize=(8, 4))
plt.title('browser_event and task relationship graph')
plt.xlabel('browser_event')
plt.ylabel('Count')
plt.legend(title='tasks')
plt.show()
```

```python
cross_tab = pd.crosstab(df['type'], df['task_subtype'])

# Plot the cross-tabulation as a stacked bar chart
cross_tab.plot(kind='bar', stacked=True, figsize=(8, 4))
plt.title('Relationship b/w tasks and task_subtypes')
plt.xlabel('types')
plt.ylabel('Count')
plt.legend(title='tasks')
plt.show()
```

```python
event_counts = df['browser_event'].value_counts()

plt.figure(figsize=(8, 4))
sns.barplot(x=event_counts.index, y=event_counts.values)
plt.title('Browser Events Frequency')
plt.xlabel('Browser_Event')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
subtype_counts = df['task_subtype'].value_counts()

plt.figure(figsize=(8, 4))
sns.barplot(x=subtype_counts.index, y=subtype_counts.values)
plt.title('Task Subtypes frequency')
plt.xlabel('Task Subtype')
plt.ylabel('Frequency')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
